```python

df=pd.read_csv('res1.csv')

vit=df[:732]
vit=vit[['timestamp','name','message','review']]
vit
```

```python
import math
vals=[]
for index,row in vit.iterrows():
    m=row['message']
    r=row['review']
    if isinstance(m,float) and isinstance(r,float):
        vals.append(np.NaN)
    else:
        vals.append(0)
vit['del']=vals
vit.dropna(subset=['del'],inplace=True)
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
vals=[]
#print(iitb)
df=vit[['message','review']]
#print(df)
for index,row in df.iterrows():
    v=row["message"]
    #print(row)
    if isinstance(v,float):
        y=row["review"]
        #print(y)
        x=re.search("(recommends)",str(y))
        #print(x)
        z=re.search("(reviewed)",str(y))
        
        if x is not None:
            vals.append(0.05)
        
        elif z is not None:
            print(z.string)
            num=re.search('[5|4|3|2|1]',z.string)
            print(num.group())
            if num.group()=='5':
                vals.append(0.3)
            elif num.group()=='4':
                vals.append(0.1)
            elif num.group()=='3':
                vals.append(0)
            elif num.group()=='2':
                vals.append(-0.1)
            elif num.group()=='1':
                vals.append(-0.3)
            #vals.append(0.05)
        if x is None and z is None:
            vals.append(-0.05)
            
        #iitb['compound']=0
        #vals.append(0)
        
    else:
        
        val=analyzer.polarity_scores(v)['compound']
        vals.append(val)
#print(vals)
vit['compound'] = vals 
```

```python
vit=vit.drop('del',axis=1)
```

```python
def sentiment(compound):
    res = ""
    if(compound >= 0.05):
        res="Positive"
    elif(compound <= - 0.05):
        res="Negative"
    else:
        res="Neutral"
    return res

vit['Sentiment'] = vit['compound'].apply(lambda x: sentiment(x))
vit
```

```python
vit['pred'] = pd.cut(vit['compound'], bins=5, labels=[1, 2, 3, 4, 5])
vals=[]
for index, row in vit.iterrows():
    if (row['compound']==0.05):
        row['pred']=4
        vals.append(4)
        
    elif row['compound']==-0.05:
        row['pred']=2
        vals.append(2)
    elif row['Sentiment']=='Positive' and row['pred']<=3:
        vals.append(4)
    elif row['Sentiment']=='Negative' and row['pred']>=3:
        vals.append(2)
    elif row['Sentiment']=='Neutral' and row['pred']!=3:
        vals.append(3)
    else:
        vals.append(row['pred'])
vit['pred']=vals    
```
