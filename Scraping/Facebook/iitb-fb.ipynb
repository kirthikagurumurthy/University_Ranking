```python
with open('res.csv') as f: 
    reader=csv.reader(f)
df=pd.read_csv('res.csv')

iitb=df[9:312]
iitb=iitb[['timestamp','name','message','review']]
    
df
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
vals=[]
#print(iitb)
df=iitb[['message','review']]
#print(df)
for index,row in df.iterrows():
    v=row["message"]
    #print(row)
    if isinstance(v,float):
        y=row["review"]
        print(y)
        x=re.search("(recommends)",str(y))
        print(x)
        if x is not None:
            vals.append(0.05)
        else:
            vals.append(-0.05)
        iitb['compound']=0
        #vals.append(0)
        
    else:
        
        val=analyzer.polarity_scores(v)['compound']
        vals.append(val)
print(vals)
iitb['compound'] = vals 
```

```python
def sentiment(compound):
    res = ""
    if(compound >= 0.05):
        res="Positive"
    elif(compound <= - 0.05):
        res="Negative"
    else:
        res="Neutral"
    return res

iitb['Sentiment'] = iitb['compound'].apply(lambda x: sentiment(x))
iitb
```

```python
iitb['pred'] = pd.cut(iitb['compound'], bins=5, labels=[1, 2, 3, 4, 5])
iitb
```
