```python
df=pd.read_csv('res2.csv')

bits=df
bits=bits[['timestamp','name','message','review']]
bits
```

```python
vals=[]
for index,row in bits.iterrows():
    m=row['message']
    r=row['review']
    if isinstance(m,float) and isinstance(r,float):
        vals.append(np.NaN)
    else:
        vals.append(0)
bits['del']=vals
bits.dropna(subset=['del'],inplace=True)
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
vals=[]
#print(iitb)
df=bits[['message','review']]
#print(df)
for index,row in df.iterrows():
    v=row["message"]
    #print(row)
    if isinstance(v,float):
        y=row["review"]
        #print(y)
        x=re.search("(recommends)",str(y))
        #print(x)
        z=re.search("(reviewed)",str(y))
        
        if x is not None:
            vals.append(0.05)
        
        elif z is not None:
            print(z.string)
            num=re.search('[5|4|3|2|1]',z.string)
            print(num.group())
            if num.group()=='5':
                vals.append(0.3)
            elif num.group()=='4':
                vals.append(0.1)
            elif num.group()=='3':
                vals.append(0)
            elif num.group()=='2':
                vals.append(-0.1)
            elif num.group()=='1':
                vals.append(-0.3)
            #vals.append(0.05)
        if x is None and z is None:
            vals.append(-0.05)
            
        #iitb['compound']=0
        #vals.append(0)
        
    else:
        
        val=analyzer.polarity_scores(v)['compound']
        vals.append(val)
#print(vals)
bits['compound'] = vals 
```

```python
bits=bits.drop('del',axis=1)
def sentiment(compound):
    res = ""
    if(compound >= 0.05):
        res="Positive"
    elif(compound <= - 0.05):
        res="Negative"
    else:
        res="Neutral"
    return res

bits['Sentiment'] = bits['compound'].apply(lambda x: sentiment(x))
bits
```

```python
def convert(score):
    OldRange = 2  
    NewRange = 4  
    NewValue = (((score + 1) * NewRange) / OldRange) + 1
    return NewValue
bits['score'] = bits['compound'].apply(lambda x: convert(x))
bits
```
