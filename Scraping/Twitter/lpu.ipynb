```python
import json
import csv
import tweepy
import re
import nltk
from nltk.stem import WordNetLemmatizer
import pandas as pd
import GetOldTweets3 as got
from nltk.corpus import stopwords
stopwords = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
import numpy as np
```

```python
lpu1=pd.read_json('lpu1.json')
lpu2=pd.read_json('lpu2.json')
```

```python
tweets_df = pd.concat([lpu1,lpu2], axis=0)
tweets_df = tweets_df.loc[:, ['text' , 'username' ,'tweet_id', 'timestamp' ,'hashtags' , 'screen_name','retweets','likes']]
tweets_df
```

```python
tweets_df.reset_index(drop=True, inplace=True)
```

```python
tweets_df = tweets_df[~tweets_df.index.duplicated()]
uni_tweets = tweets_df.loc[tweets_df['screen_name'].isin(['lpuuniversity','OfficialLPUBats','LPUyfact'])]
```

```python
uni_tweets['uni_handle'] = 1
uni_tweets
```

```python
tweets_df = tweets_df[~tweets_df.index.duplicated()]
tweets_df['uni_handle'] = uni_tweets['uni_handle']
lpu = tweets_df.combine_first(tweets_df)
lpu
```

```python
lpu['uni_handle'] = lpu['uni_handle'].replace(np.nan, 0)
lpu[lpu['uni_handle'] == 1]
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
lpu['compound'] = [analyzer.polarity_scores(v)['compound'] for v in lpu['text']]
```

```python
def sentiment(compound):
    res = ""
    if(compound >= 0.05):
        res="Positive"
    elif(compound <= - 0.05):
        res="Negative"
    else:
        res="Neutral"
    return res

lpu['Sentiment'] = lpu['compound'].apply(lambda x: sentiment(x))
```

```python
lpu['score'] = pd.cut(lpu['compound'], bins=5, labels=[1, 2, 3, 4, 5])
lpu
```

```python
lpu.to_csv('../../Datasets/final_lpu_withlikes.csv')
```
