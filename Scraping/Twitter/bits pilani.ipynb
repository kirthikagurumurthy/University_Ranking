```python
text_query = 'BITS Pilani'
since_date = '2019-06-01'
# Creation of query object
tweetCriteria = got.manager.TweetCriteria().setQuerySearch(text_query).setSince(since_date).setLang('en')
# Creation of list that contains all tweets
tw1 = got.manager.TweetManager.getTweets(tweetCriteria)
# Creating list of chosen tweet data
text_tweets1 = [[tweet.text ,tweet.date,tweet.username,tweet.retweets,tweet.hashtags] for tweet in tw1]
```

```python
text_query = 'Birla Institute of Technology Pilani'
since_date = '2019-06-01'
# Creation of query object
tweetCriteria = got.manager.TweetCriteria().setQuerySearch(text_query).setSince(since_date).setLang('en')
# Creation of list that contains all tweets
tw2 = got.manager.TweetManager.getTweets(tweetCriteria)
# Creating list of chosen tweet data
text_tweets2 = [[tweet.text ,tweet.date,tweet.username,tweet.retweets,tweet.hashtags] for tweet in tw2]
```

```python
len(text_tweets2)
```

```python
tweets_df = pd.DataFrame(text_tweets1+text_tweets2)
tweets_df.columns = ['Tweet_text' , 'date' , 'username','retweets','hashtags']
tweets_df = tweets_df.drop_duplicates()
tweets_df
```

```python
uni_tweets = tweets_df.loc[tweets_df['username'].isin(['bitspilaniindia','BITSApogee','BITSPilaniDubai','BITSPilaniGoa','BITS_Pilani','celbitsgoa','ARCDBITSPilani'])]
uni_tweets['uni_handle'] = 1
uni_tweets
```

```python
tweets_df['uni_handle'] = uni_tweets['uni_handle']
bitsp = tweets_df.combine_first(tweets_df)
bitsp
```

```python
bitsp[bitsp['uni_handle'] == 1]
```

```python
bitsp['uni_handle'] = bitsp['uni_handle'].replace(np.nan, 0)
bitsp_final = bitsp.drop_duplicates()
bitsp_final
```

```python
#Finding sentiment of the tweets using VADER
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
bitsp['compound'] = [analyzer.polarity_scores(v)['compound'] for v in bitsp['Tweet_text']]
bitsp
```

```python
def sentiment(compound):
    res = ""
    if(compound >= 0.05):
        res="Positive"
    elif(compound <= - 0.05):
        res="Negative"
    else:
        res="Neutral"
    return res

bitsp['Sentiment'] = bitsp['compound'].apply(lambda x: sentiment(x))
```
