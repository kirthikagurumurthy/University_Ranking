```python
import json
import csv
import tweepy
import re
import nltk
from nltk.stem import WordNetLemmatizer
import pandas as pd
import GetOldTweets3 as got
from nltk.corpus import stopwords
stopwords = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
import numpy as np
```

```python
bits1=pd.read_json('bitsp2019.json')
bits2=pd.read_json('bitsp2020.json')
```

```python
tweets_df = pd.concat([bits1,bits2], axis=0)
tweets_df = tweets_df.loc[:, ['text' , 'username' ,'tweet_id', 'timestamp' ,'hashtags' , 'screen_name','retweets','likes']]
tweets_df
```

```python
tweets_df.reset_index(drop=True, inplace=True)
```

```python
tweets_df = tweets_df[~tweets_df.index.duplicated()]
uni_tweets = tweets_df.loc[tweets_df['screen_name'].isin(['bitspilaniindia','BITSApogee','BITSPilaniDubai','BITSPilaniGoa','BITS_Pilani','celbitsgoa','ARCDBITSPilani'])]
```

```python
uni_tweets['uni_handle'] = 1
uni_tweets
```

```python
tweets_df = tweets_df[~tweets_df.index.duplicated()]
tweets_df['uni_handle'] = uni_tweets['uni_handle']
bitsp = tweets_df.combine_first(tweets_df)
bitsp
```

```python
bitsp['uni_handle'] = bitsp['uni_handle'].replace(np.nan, 0)
bitsp[bitsp['uni_handle'] == 1]
```

```python
#Finding sentiment of the tweets using VADER
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
bitsp['compound'] = [analyzer.polarity_scores(v)['compound'] for v in bitsp['text']]
bitsp
```

```python
def sentiment(compound):
    res = ""
    if(compound >= 0.05):
        res="Positive"
    elif(compound <= - 0.05):
        res="Negative"
    else:
        res="Neutral"
    return res

bitsp['Sentiment'] = bitsp['compound'].apply(lambda x: sentiment(x))
```

```python
bitsp['score'] = pd.cut(bitsp['compound'], bins=5, labels=[1, 2, 3, 4, 5])
bitsp
```
