```python
df= pd.read_csv('../../Datasets/Google/srm-gm-reviews.csv')
df
```

```python
df=df[['id_review','rating']]
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
vals=[]
for index,row in df.iterrows():
    v=row['id_review']
    if isinstance(v,float):
        vals.append(1.5)
    else:
        val=analyzer.polarity_scores(v)['compound']
        vals.append(val)
df['compound']=vals
```

```python
def sentiment(compound):
    res = ""
    if(compound >= 0.05):
        res="Positive"
    elif(compound <= - 0.05):
        res="Negative"
    else:
        res="Neutral"
    return res

df['Sentiment'] = df['compound'].apply(lambda x: sentiment(x))
df
```

```python
df['pred'] = pd.cut(df['compound'], bins=5, labels=[1, 2, 3, 4, 5])

vals=[]
for index,row in df.iterrows():
    if isinstance(row['id_review'],float):
        row['pred']=row['rating']
        vals.append(row['pred'])
    else:
        row['fin_pred']= (row['pred']+row['rating'])/2
        vals.append((row['pred']+row['rating'])/2)
df['temp_pred']=vals
df
```

```python
df['fin_pred'] = pd.cut(df['temp_pred'], bins=5, labels=[1, 2, 3, 4, 5])

```
